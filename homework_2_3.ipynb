```

import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn import datasets
# %matplotlib inline

diabetes = datasets.load_diabetes()
X = diabetes.data
y = diabetes.target

row, col = X.shape

b = np.ones(row)
X = np.insert(X, 0, values=b, axis=1)
print(X)

col += 1

def cost_function(X, Y, B):
 length = len(Y)
 res = np.sum((X.dot(B)-Y)**2)/(2*length)
 return res

def batch_gradient_descent(X,Y,B,alpha,epochs):
  cost_history = [0] * epochs
  leng = len(Y)

  for epoch in range(epochs):
    h = X.dot(B)
    loss = h - Y
    gradient = X.T.dot(loss) / leng
    B = B - alpha * gradient
    cost = cost_function(X,Y,B)
    cost_history[epoch] = cost_function

  return B, cost_history


B = np.zeros(X.shape[1])
alpha = 0.08
epo = 20000
newB, cost_history = batch_gradient_descent(X, y, B, alpha, epo)
print(newB)

def pred(X,newB):
    return X.dot(newB)

y_ = pred(X,newB)
print(y_)

RSS = np.mean((y_ - y)**2)/(np.std(y)**2)
Rsq = 1-RSS
print("RSS per sample = {0:f}".format(RSS))
print("R^2 =            {0:f}".format(Rsq))
```
